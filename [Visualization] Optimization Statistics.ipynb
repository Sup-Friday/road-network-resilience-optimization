```python
import csv
import numpy as np
import matplotlib.pyplot as plt

folder = 'data_processed/csv'

CC_connected_components_statistics_dir = folder + '/minimizing_connected_components_statistics_of_connected_components.csv'
CC_recovered_trips_statistics_dir = folder + '/minimizing_connected_components_statistics_of_recovered_trips.csv'
IT_connected_components_statistics_dir = folder + '/minimizing_infeasible_trips_statistics_of_connected_components.csv'
IT_recovered_trips_statistics_dir = folder + '/minimizing_infeasible_trips_statistics_of_recovered_trips.csv'

# Different budgets list
budgets = np.arange(11000, step=1000)

# Connceted components list by Minimizing connected components (CC)
CC_connected_components = []
with open(CC_connected_components_statistics_dir) as f:
    datareader = csv.DictReader(f)
    for row in datareader:
        data = list(row.items())
        CC_connected_components.append(int(float(data[1][1])))
print('Connceted components by minimizing CC:', CC_connected_components)

# Recovered trips list by Minimizing infeasible trips (IT)
IT_connected_components = []
with open(IT_connected_components_statistics_dir) as f:
    datareader = csv.DictReader(f)
    for row in datareader:
        data = list(row.items())
        IT_connected_components.append(int(float(data[1][1])))
print('Connceted components by minimizing IT:', IT_connected_components)

# Recovered trips list by Minimizing connected components (CC)
CC_recovered_trips = []
with open(CC_recovered_trips_statistics_dir) as f:
    datareader = csv.DictReader(f)
    for row in datareader:
        data = list(row.items())
        CC_recovered_trips.append(int(float(data[2][1])))
print('Recovered trips by minimizing CC:', CC_recovered_trips)

# Recovered trips list by Minimizing infeasible trips (IT)
IT_recovered_trips = []
with open(IT_recovered_trips_statistics_dir) as f:
    datareader = csv.DictReader(f)
    for row in datareader:
        data = list(row.items())
        IT_recovered_trips.append(int(float(data[2][1])))
print('Recovered trips by minimizing IT:', IT_recovered_trips)
```

```python
# Show results of connected components given only Minimizing CC
fig, ax = plt.subplots()
fig.set_size_inches(10, 7)

ax.plot(budgets, CC_connected_components, linewidth = '3', color='#f88379', label='minimizing number of connected components')
ax.set_xlabel('Budget (meters)', fontsize=18)
ax.set_ylabel('Number of connected components', fontsize=18)

plt.legend(loc='upper right', fontsize=18)
plt.style.use(plt.style.available[7])
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
# Show results of connected components given both Minimizing CC and IT 
fig, ax = plt.subplots()
fig.set_size_inches(10, 7)

ax.plot(budgets, IT_connected_components, linewidth = '3', color = '#1f77b4', label='minimizing number of infeasible trips')
ax.plot(budgets, CC_connected_components, linewidth = '3', color='#f88379', label='minimizing number of connected components')
ax.set_xlabel('Budget (meters)', fontsize=18)
ax.set_ylabel('Number of connected components', fontsize=18)

plt.legend(loc='upper right', fontsize=18)
plt.style.use(plt.style.available[7])
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
# Show results of recovered trips given only Minimizing IT
fig, ax = plt.subplots()
fig.set_size_inches(10, 7)

ax.plot(budgets, IT_recovered_trips, linewidth = '3', color = '#1f77b4', label='minimizing number of infeasible trips')
ax.set_xlabel('Budget (meters)', fontsize=18)
ax.set_ylabel('Number of recovered trips', fontsize=18)

plt.style.use(plt.style.available[7])
plt.legend(loc='lower right', fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
# Show results of recovered trips given both Minimizing IT and CC
fig, ax = plt.subplots()
fig.set_size_inches(10, 7)

ax.plot(budgets, CC_recovered_trips, linewidth = '3', color='#f88379', label='minimizing number of connected components')
ax.plot(budgets, IT_recovered_trips, linewidth = '3', color='#1f77b4', label='minimizing number of infeasible trips')
ax.set_xlabel('Budget (meters)', fontsize=18)
ax.set_ylabel('Number of recovered trips', fontsize=18)

plt.style.use(plt.style.available[7])
plt.legend(loc='lower right', fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
